```python
def check_feat(feature_name):
    '''
    This is a helper function that provide basic
    information about a specific feature
    
    INPUT: feature name of the dataset to look
    OUTOUT: It will print info about: 
        first 7 elements, 
        numbers of NaNs , 
        datatype and 
        unique values    
    '''
    print(df[feature_name].head(7))
    print(f'Number of NaN is : {df[feature_name].isnull().sum()}' )
    print(f'Datatype : {type(df[feature_name][0])}')
    print(df[feature_name].unique())
    
```

```python
check_feat('room_type')
```

```python
#Exploring non numerical values
df.describe(include=[np.object]).T
```

```python
#Exploring numerical values
df.describe(include=[np.number]).T
```

```python
#Checking for NaNs and printing the features above the threlshod
feature = []
perc_nulls= []
for col in df.columns:
    if df[col].isnull().sum() > 0.3*df.shape[0]:
        feature.append(col)
        perc_nulls.append(df[col].isnull().sum()/df.shape[0]*100)
        print(col, df[col].isnull().sum()/df.shape[0]*100)
    
```

```python
#Dropping the features from the NaN analisys
for col in feature:
    df.drop([col],axis=1,inplace=True)
```

```python
#Visualizing the features with more than 30% of nan values
ds = {'Feature': feature , 'Percentage':perc_nulls}
dx = pd.DataFrame(ds)
sb.barplot(y='Feature', x='Percentage', data=dx)
```

Checking the format and type of "price" feature

```python
#The "price" fatures is string type 
check_feat('price')
```

Prices are strings and have a weird format, we have to remove the $ sign and the "00" at the end and conver
it to float.

```python
df['price'] = df['price'].apply(lambda x: x[1:-3]).apply(lambda x: x.replace(',','')).astype(np.int64)
```

```python
df['price'].head()
```

Visualize the "price" feature and check if there are some outliers. If so we will use the [Tukey Rule](https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/box-whisker-plots/a/identifying-outliers-iqr-rule) 
 to detect them. 

```python
print(df['price'].describe()) 
df['price'].plot(kind ='box')
plt.savefig('checkprices.png', dpi=300)
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

IQR = Q3 -Q1

Max = Q3 + 1.5*IQR
Min = Q1 - 1.5*IQR

print('Min value {} , Max value {}'.format(Min,Max))
print(f'Total count for price higher than {Max} = ' + str(df[df['price'] > Max]['price'].count()))
print('A look at the outlier prices :' + str(np.sort(df[df['price'] > Max]['price'].unique())))
```

```python
#Filtering the dataframe by price using the Max value we obtained after Tukey rule
df=df[df.price<Max]
```
